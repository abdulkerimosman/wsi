```python
data_dict = {"Date": date_columns}
for school, data in zip(schools, random_data):
    data_dict[school] = data

df = pd.DataFrame(data_dict)

df
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt
```

```python
# Ensure the Date column is in datetime format and set as index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df = df.asfreq('QS')  # Explicitly set quarterly frequency
```

```python
# List of all school columns
school_columns = df.columns
```

```python
# Forecast horizon
forecast_steps = 4
forecast_start = '2025-01-01'
forecast_index = pd.date_range(start=forecast_start, periods=forecast_steps, freq='QS')

# Create a new DataFrame for forecasts
forecast_df = pd.DataFrame(index=forecast_index)

# Loop through each school column
for school in school_columns:
    print(f"Processing: {school}")
    
    # Extract data for the current school
    school_data = df[school]
    
    # Define SARIMA model with adjusted parameters
    try:
        model = SARIMAX(school_data,
                        order=(1, 1, 1),  # p, d, q
                        seasonal_order=(0, 0, 0, 4),  # Simplified seasonal terms
                        enforce_stationarity=False,
                        enforce_invertibility=False)
        
        # Fit the model
        results = model.fit(disp=False)
        
        # Forecast for the specified steps ahead
        forecast = results.get_forecast(steps=forecast_steps)
        forecast_values = forecast.predicted_mean
        
        # Round forecast values down to the nearest integer
        forecast_values = np.floor(forecast_values).astype(int)
        
        # Add forecast to the forecast DataFrame
        forecast_df[school] = forecast_values
    except Exception as e:
        print(f"Error processing {school}: {e}")
        forecast_df[school] = np.nan  # Fill with NaN in case of errors

# Join the forecast DataFrame to the original DataFrame
df_combined = pd.concat([df, forecast_df])
```
